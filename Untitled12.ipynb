```python
import cmath

# 1. Define a Python function that takes coefficients a and b as input and returns the solution for a linear equation (ax + b = 0). Handle the case where 'a' is zero.
def solve_linear(a, b):
    """Solves a linear equation of the form ax + b = 0."""
    if a == 0:
        if b == 0:
            return "Infinite solutions"
        else:
            return "No solution"
    else:
        return -b / a

# 2. Define a Python function that takes coefficients a, b, and c as input and returns the solution(s) for a quadratic equation (ax^2 + bx + c = 0) using the quadratic formula. This function should be able to return both real and complex solutions.
def solve_quadratic(a, b, c):
    """Solves a quadratic equation of the form ax^2 + bx + c = 0."""
    delta = (b**2) - 4*(a*c)

    if delta >= 0:
        # Real solutions
        x1 = (-b - cmath.sqrt(delta)) / (2*a)
        x2 = (-b + cmath.sqrt(delta)) / (2*a)
        return x1, x2
    else:
        # Complex solutions
        x1 = (-b - cmath.sqrt(delta)) / (2*a)
        x2 = (-b + cmath.sqrt(delta)) / (2*a)
        return x1, x2

# 4. Call the linear equation solving function with sample coefficients and print the result.
print("Solving linear equation 2x + 4 = 0:")
print("Solution:", solve_linear(2, 4))

print("\nSolving linear equation 0x + 5 = 0:")
print("Solution:", solve_linear(0, 5))

print("\nSolving linear equation 0x + 0 = 0:")
print("Solution:", solve_linear(0, 0))

# 5. Call the quadratic equation solving function with sample coefficients that result in real solutions and print the results.
print("\nSolving quadratic equation x^2 - 5x + 6 = 0 (Real solutions):")
sol1_real, sol2_real = solve_quadratic(1, -5, 6)
print("Solutions:", sol1_real, sol2_real)

# 6. Call the quadratic equation solving function with sample coefficients that result in complex solutions and print the results.
print("\nSolving quadratic equation x^2 + x + 1 = 0 (Complex solutions):")
sol1_complex, sol2_complex = solve_quadratic(1, 1, 1)
print("Solutions:", sol1_complex, sol2_complex)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 2. Define a function linear_func(x, a, b) that calculates a*x + b.
def linear_func(x, a, b):
    """Calculates the value of a linear function ax + b."""
    return a * x + b

# 3. Define a function quadratic_func(x, a, b, c) that calculates a*x**2 + b*x + c.
def quadratic_func(x, a, b, c):
    """Calculates the value of a quadratic function ax^2 + bx + c."""
    return a * x**2 + b * x + c

# 4. Create an array of x-values using np.linspace for plotting the functions.
x_values_linear = np.linspace(-5, 5, 100)
x_values_quadratic_real = np.linspace(0, 4, 100)
x_values_quadratic_complex = np.linspace(-2, 2, 100)


# 5. Plot the linear function y = 2x + 4. Label the x-axis as 'x', the y-axis as 'y', and add a title 'Linear Equation: 2x + 4 = 0'. Add a horizontal line at y=0 to show the root.
plt.figure(figsize=(8, 6))
plt.plot(x_values_linear, linear_func(x_values_linear, 2, 4), label='2x + 4')
plt.axhline(0, color='red', linestyle='--', label='y=0 (Root)')
plt.title('Linear Equation: 2x + 4 = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# 6. Plot the quadratic function y = x^2 - 5x + 6. Label the x-axis as 'x', the y-axis as 'y', and add a title 'Quadratic Equation: x^2 - 5x + 6 = 0'. Add a horizontal line at y=0 to show the roots.
plt.figure(figsize=(8, 6))
plt.plot(x_values_quadratic_real, quadratic_func(x_values_quadratic_real, 1, -5, 6), label='x^2 - 5x + 6')
plt.axhline(0, color='red', linestyle='--', label='y=0 (Roots)')
plt.title('Quadratic Equation: x^2 - 5x + 6 = 0')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# 7. Plot the quadratic function y = x^2 + x + 1. Label the x-axis as 'x', the y-axis as 'y', and add a title 'Quadratic Equation: x^2 + x + 1 = 0'. Add a horizontal line at y=0.
plt.figure(figsize=(8, 6))
plt.plot(x_values_quadratic_complex, quadratic_func(x_values_quadratic_complex, 1, 1, 1), label='x^2 + x + 1')
plt.axhline(0, color='red', linestyle='--', label='y=0')
plt.title('Quadratic Equation: x^2 + x + 1 = 0 (No Real Roots)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
